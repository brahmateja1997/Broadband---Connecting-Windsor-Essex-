```python

# Define a function 'plot_cat' that takes in a dataframe, x column name, y column name and title for the plot
def plot_cat(df, x, y, title):
    
    # Setting the size of the figure using matplotlib
    plt.figure(figsize=(12,7))
    
    # Grouping the dataframe by the x column and calculating the mean of the y column for each group
    cat_df = df.groupby([x])[y].mean().sort_values(ascending=False).reset_index()
    
    # Plotting a bar chart using seaborn
    sns.barplot(x=x, y=y, data=cat_df, palette="cubehelix")
    
    # Rotating the x-axis tick labels by 90 degrees to avoid overlapping
    plt.xticks(rotation=90)
    
    # Setting the title of the chart
    plt.title(title, fontsize=20)
    
    # Displaying the plot
    plt.show()

# Calling the 'plot_cat' function twice with different parameters for two separate plots
plot_cat(df, 'isp_name', 'download_mbps', 'Average Download Speed w.r.t Internet service providers')
plot_cat(df, 'isp_name', 'upload_mbps', 'Average Upload Speed w.r.t Internet service providers')
```

We can observe that Cogeco has the higher average download speed while Rogers provides higher upload speed.

```python
# Converting RainType to categorical variable to assign categories
dfw['RainType'] = pd.Categorical(dfw['RainType'])

# Creating a scatter plot with RainType on the x-axis and Jitter on the y-axis using plt.scatter() function of matplotlib library.
scatter = plt.scatter(dfw['RainType'], dfw['Jitter'], c=dfw["RainType"].cat.codes, cmap='rainbow')

# Adding a label to x-axis using .xlabel() function.
plt.xlabel('Rain Type')

# Adding a label to y-axis using .ylabel() function.
plt.ylabel('Jitter')

# Adding a title to the plot using .title() function.
plt.title('Jitter vs Rain Type')

# Adding a color bar 
plt.colorbar(scatter)

# Displaying the plot using .show() function.
plt.show()


```

```python
# Converting RainType to categorical variable to assign categories
dfw['RainType'] = pd.Categorical(dfw['RainType'])

# Creating a scatter plot with RainType on the x-axis and Jitter on the y-axis using plt.scatter() function of matplotlib library.
scatter = plt.scatter(dfw['RainType'], dfw['Ping (ms)'], c=dfw["RainType"].cat.codes, cmap='tab10', s=100, alpha=0.8, edgecolors='black')

# Adding a label to x-axis using .xlabel() function.
plt.xlabel('Rain Type')

# Adding a label to y-axis using .ylabel() function.
plt.ylabel('Ping')

# Adding a title to the plot using .title() function.
plt.title('Ping vs Rain Type')

# Adding a color bar 
plt.colorbar(scatter)

# Customizing the scatter plot markers
scatter.set_facecolor("white")
scatter.set_edgecolor('black')

# Displaying the plot using .show() function.
plt.show()
```

Hence it is observable that as rain increases Ping also increases.

```python
# Converting SnowType to categorical variable to assign categories
dfw['SnowType'] = pd.Categorical(dfw['SnowType'])

# Creating a scatter plot with Ping (ms) on the x-axis and SnowType on the y-axis using plt.scatter() function of matplotlib library.
scatter = plt.scatter(dfw['Ping (ms)'], dfw['SnowType'], c=dfw["SnowType"].cat.codes, cmap='Set2', s=100, alpha=0.8, edgecolors='black')

# Adding a label to x-axis using .xlabel() function.
plt.xlabel('Ping')

# Adding a label to y-axis using .ylabel() function.
plt.ylabel('Snow Type')

# Adding a title to the plot using .title() function.
plt.title('Ping vs Snow Type')

# Adding a color bar 
plt.colorbar(scatter)

# Customizing the scatter plot markers
scatter.set_facecolor("white")
scatter.set_edgecolor('black')

# Inverting the y-axis
plt.gca().invert_yaxis()

# Setting the figure size
fig = plt.gcf()
fig.set_size_inches(8, 6)

# Displaying the plot using .show() function.
plt.show()
```
